##Нормальний розподіл

Нормальний розподіл (Гауссівський розподіл)
Це симетричний розподіл, де більшість значень зосереджені навколо середнього значення. Характеризується двома параметрами: середнім (μ) і стандартним відхиленням (σ).
Нижче наведений код для створення набору даних нормального розподілу.

---

- mu - середнє значення
- sigma - стандартне відхилення




```python
mu, sigma = 0, 1
normal = np.random.normal(mu, sigma, 1000)

count, bins, ignored = plt.hist(normal, 30, density=True, alpha=0.6, color='g')

plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp(- (bins - mu)**2 / (2 * sigma**2)), linewidth=2, color='r')

plt.title('Нормальний розподіл')
plt.xlabel('Значення')
plt.ylabel('Ймовірність')

plt.grid(True)
plt.show()
```

###Висновок
На графіку представлено приклад нормального розподілу. Зеленим кольором зображено гістограму вибірки даних, яка апроксимується нормальним розподілом, а червоною лінією - теоретичну криву нормального розподілу з заданими параметрами середнього значення (μ = 0) та стандартного відхилення (σ = 1). Цей графік ілюструє, як дані розподілені навколо середнього значення, формуючи характерну "дзвоноподібну" криву.

##Рівномірний розподіл

Рівномірний розподіл Усі події мають однакову ймовірність виникнення. Має два параметри: мінімум (a) і максимум (b).
Нижче наведений код для створення набору даних рівномірного розподілу.

---
- a - min
- b - max


```python
a, b = -2, 2
uniform = np.random.uniform(a, b, 1000)

count, bins, ignored = plt.hist(uniform, 30, density=True, alpha=0.6, color='b')
plt.plot(bins, np.ones_like(bins) / (b - a), linewidth=2, color='r')
plt.title('Рівномірний розподіл')
plt.xlabel('Значення')
plt.ylabel('Ймовірність')
plt.grid(True)
plt.show()
```

###Висновок
На графіку ілюструється рівномірний розподіл. Значення зразка, отримані з рівномірного розподілу між a = -2 та  b = 2, зображені синім кольором у вигляді гістограми. Червона лінія показує теоретичну рівну ймовірність для всіх значень у цьому інтервалі. Це демонструє, що кожне значення всередині вказаного інтервалу має однакову ймовірність виникнення, що є характерною рисою рівномірного розподілу.

##Експоненційний розподіл

Експоненційний розподіл є важливим розподілом у теорії ймовірностей та статистики, який часто застосовується для моделювання часу між наступними подіями в процесі, що відбувається з постійною середньою інтенсивністю. Ключовий параметр - λ, середнє число подій у одиницю часу.
Нижче наведений код для створення набору даних експонеційного розподілу.

---

- lam - cереднє число подій у одиницю часу

```python
lam = 0.5
exponential = np.random.exponential(1/lam, 1000)

count, bins, ignored = plt.hist(exponential, 30, density=True, alpha=0.6, color='orange')
plt.plot (bins, lam * np.exp(-lam * bins), linewidth=2, color='r')
plt.title( 'Експоненційний розподіл')
plt.xlabel ( 'Значення')
plt.ylabel ( 'Ймовірність')
plt.grid(True)
plt.show()
```

###Висновок
На графіку вище демонструється експоненційний розподіл. Дані, отримані з експоненційного розподілу, зображені помаранчевим кольором у вигляді гістограми. Червона лінія показує теоретичну криву щільності ймовірності для експоненційного розподілу з параметром λ=0.5. Графік відображає характерний спадаючий тренд, що вказує на зменшення ймовірності великих значень у вибірці, що є типовим для експоненційного розподілу.

##Пуасонівський розподіл

Описує кількість подій, що відбуваються за фіксований інтервал часу або в певній області простору, при постійній середній швидкості подій. Ключовий параметр - λ, середня кількість подій на інтервал.
Нижче наведений код для створення набору даних пуасонівського розподілу.

---

- lam_p - середня кількість подій на інтервал

```python
lam_p = 3
poisson = np.random.poisson(lam_p, 1000)

plt.hist(poisson, bins=range(min(poisson), max(poisson) + 1, 1), density=True, alpha=0.6, color='purple')
plt.title('Пуассонівський розподіл')
plt.xlabel ('Кількість подій')
plt.ylabel ('Ймовірність')
plt.grid (True)
plt.show()
```

###Висновок
На представленому графіку ілюструється Пуассонівський розподіл. Дані, сгенеровані з Пуассонівського розподілу з параметром λ=3, що вказує на середню кількість подій на інтервал, показані фіолетовим кольором у вигляді гістограми. Графік показує ймовірність спостереження різної кількості подій, демонструючи характерну для Пуассонівського розподілу форму, де висота кожного стовпця відповідає ймовірності спостереження певної кількості подій.

##Імпорт додаткового модуля

```python
from scipy.stats import shapiro
```

Опис: Тест Шапіро-Вілка є одним із найбільш потужних тестів для перевірки гіпотези про нормальність розподілу вибірки. Він особливо корисний для малих до середніх вибірок і часто використовується для аналізу статистичних даних у різних галузях науки.

Переваги/недоліки:
- Переваги: Вважається одним з найбільш потужних критеріїв для перевірки нормальності.
- Недоліки: Ефективність знижується при дуже великих вибірках.

Гіпотези:
- Нульова гіпотеза (H0): Вибірка походить з нормального розподілу.
- Альтернативна гіпотеза (H1): Вибірка не походить з нормального розподілу.

Так як недоліки цього тесту у великих вибірках, зменшимо наші вибірки вибраних розподілів

---
- new_normal - нормальний
- new_exponential - експоненційний
- new_binomial - біноміальний


```python
new_normal = np.random.normal(mu, sigma, 100)
new_exponential = np.random.exponential(1/lam, 100)
new_binomial = np.random.binomial(n, p, 100)

datasets = {
    "Нормальний розподіл": new_normal,
    "Експоненційний розподіл": new_exponential,
    "Біноміальний розподіл": new_binomial
}

alpha = 0.05  # Рівень значущості

# Перевірка нормальності за допомогою тесту Шапіро-Вілка
results = {}
for name, data in datasets.items():
    stat, p_value = shapiro(data)
    results[name] = (stat, p_value)

# Оцінка результатів тесту
hypothesis_results = {}
for name, (stat, p_value) in results.items():
    if p_value > alpha:
        hypothesis_results[name] = "Нульову гіпотезу не відхилено - розподіл може бути нормальним."
    else:
        hypothesis_results[name] = "Нульову гіпотезу відхилено - розподіл не є нормальним."

# Виведення оцінки результатів тестів
print("Оцінка результатів тесту:")
for name, result in hypothesis_results.items():
    print(f"{name}: {result}")
    print("----------------------------")

# Виведення результатів тестів
for name, (stat, p_value) in results.items():
    print(f"Результати тесту для {name}:")
    print(f"Статистика тесту: {stat}")
    print(f"p-значення: {p_value}")
    print()
```
