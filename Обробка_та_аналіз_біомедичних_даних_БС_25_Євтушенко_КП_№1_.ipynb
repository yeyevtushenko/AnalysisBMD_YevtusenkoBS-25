##Робота зі зміненням завантаженого масиву

###Створення новиго масиву даних, що містить лише перші 10 рядків з завантаженого масиву

```python
cut_diabetes = diabetes.iloc[0:10, :]
cut_diabetes
```

###Додавання нового стовпчика та заповнення його даними, що відносять кожний з рядків до однієї з двох груп.

```python
cut_diabetes["Group"] = cut_diabetes["Age"].apply(lambda age: "Young" if age < 50 else "Old")
cut_diabetes
```

###Виведення статистики з розділенням на групи

```python
cut_diabetes.groupby("Group").describe()
```

##Збереження нових даних

```python
cut_diabetes.to_csv("cut_diabetes.csv", encoding= "utf8")
```
